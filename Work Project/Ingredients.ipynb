```python
import os
import codecs
import jieba
from jieba import analyse
import pandas as pd
from pandas import Series as sr, DataFrame as df  ##从pandas包引入Series与DataFrame格式
from collections import Counter as cr  ##引入Counter进行计数
import jieba.posseg as pseg  ##引入结巴分词词性标注
```

```python
file = pd.read_csv('CodingIngredients.csv', header=0,encoding='gb18030',dtype=str)
```

```python
rows = file[['pname','Ingredients']]
rows.head(5)
```

```python
def stopwordslist(filepath):  
    stopwords = [line.strip() for line in open(filepath, 'r', encoding='gb18030').readlines()]  
    return stopwords  
def seg_sentence(sentence):  
    sentence_seged = jieba.cut(sentence.strip())  
    stopwords = stopwordslist('stopwords_pname_ansi.txt')  # 这里加载停用词的路径  
    outstr = ''  
    for word in sentence_seged:  
        if word not in stopwords:  
            if word != '\t':  
                outstr += word  
                outstr += " "  
    return outstr 
```

```python
info=[]
for i in range(1000):
    words = jieba.analyse.extract_tags(seg_sentence(str(rows['Ingredients'][i])+str(rows['pname'][i])), topK=10)
    info.append(words) 
m =pd.DataFrame(info)  
m.to_csv("coding.csv",encoding='utf-8_sig')
```

```python
pd.concat([rows,df(info)],axis=1)
```
